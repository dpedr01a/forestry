```python
### Points mode
points_gdf = gpd.GeoDataFrame()
for i in range(len(nishi)):
    polygon = nishi[i:i+1]
    if polygon.iat[0,polygon.columns.get_loc('Density60')] == 0:
        continue
    grid_res = 100 / round(math.sqrt(polygon.iat[0,polygon.columns.get_loc('Density60')]))
    age = int(polygon.iat[0,polygon.columns.get_loc('Age60')])
    age = age if age < 200 else 200
    
    minx = np.floor(polygon.geometry.bounds.iat[0,0] / grid_res) * grid_res
    maxx = np.ceil(polygon.geometry.bounds.iat[0,2] / grid_res) * grid_res
    miny = np.floor(polygon.geometry.bounds.iat[0,1] / grid_res) * grid_res
    maxy = np.ceil(polygon.geometry.bounds.iat[0,3] / grid_res) * grid_res

    x = np.arange(minx,maxx+grid_res/2,grid_res)
    y = np.arange(miny,maxy+grid_res/2,grid_res)
    XX,YY = np.meshgrid(x,y)

    # use shapely.vectorized to find all points within the polygon
    in_polygon = shapely.vectorized.contains(polygon.iat[0,polygon.columns.get_loc('geometry')],XX,YY)

    # filter the points and flatten them to 1D vectors
    x_in_polygon = XX[in_polygon].ravel()
    y_in_polygon = YY[in_polygon].ravel()
    point_geom = [Point(x,y) for x,y in zip(x_in_polygon,y_in_polygon)]
    ages = [int(age)]*len(point_geom)
    point_gdf = gpd.GeoDataFrame({'tree':ages},geometry=point_geom,crs=polygon.crs)
    points_gdf = gpd.GeoDataFrame(pd.concat([points_gdf,point_gdf],axis=0),crs=polygon.crs)
```

```python
points_gdf['tree'] = 'Age'+points_gdf['tree'].astype(int).astype(str)
points_gdf = points_gdf.reset_index(drop=True)
points_gdf.insert(0,'id',points_gdf.index+1)
points_gdf = points_gdf.to_crs('epsg:4326')
points_gdf
```
